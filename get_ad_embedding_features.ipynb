```python
import pandas as pd
import gc
```

```python
ad_id_embedding = pd.read_csv('output/model/ad_id_embedding.csv')
ad_id_embedding
```

训练集

```python
train_ad = pd.read_csv('data/train_preliminary/ad.csv')
train_click_log = pd.read_csv('data/train_preliminary/click_log.csv')
train = pd.merge(train_ad, train_click_log, how='outer')
del train_ad, train_click_log
gc.collect()
train
```

```python
train_features = train[['user_id', 'ad_id']]
del train
gc.collect()
train_features
```

```python
train_id = train_features[['user_id']].drop_duplicates()
train_id
```

```python
train_ad_id_embedding = pd.merge(train_features, ad_id_embedding, on='ad_id', how='left')
train_ad_id_embedding
```

```python
def extract_feat(df_, key, name, L, train_id):
    for i in range(L):
        t = df_.groupby(key).aggregate({'{0}_embe_{1}'.format(name, i):['mean','min','max']}).reset_index()
        t.columns = ['_'.join(j).strip('_') for j in t.columns.values]
        train_id = pd.merge(train_id, t, on=key, how='left')
        del t 
        gc.collect()
    return train_id
```

```python
train_ad_id_embedding_df= extract_feat(train_ad_id_embedding, 'user_id', 'ad_id', 32, train_id)
train_ad_id_embedding_df
```

```python
train_ad_id_embedding_df.to_csv('output/features/train/ad_id_embedding_df.csv', index=False)
```

```python
del train_features, train_id, train_ad_id_embedding, train_ad_id_embedding_df
gc.collect()
```
