```python
import pandas as pd
import gc
```

```python
#读入整个的embedding
creative_id_embedding = pd.read_csv('output/model/creative_id_embedding.csv')
creative_id_embedding
```

训练集

```python
train_click_log = pd.read_csv('data/train_preliminary/click_log.csv')
train_click_log 
# test_click_log = pd.read_csv('data/test/click_log.csv')
# train_test_click_log = train_click_log.append(test_click_log)
# train_test_features = train_test_click_log[['user_id', 'creative_id']]
# train_test_features
```

```python
train_features = train_click_log[['user_id', 'creative_id']]
del train_click_log
gc.collect()
train_features
```

```python
train_id = train_features[['user_id']].drop_duplicates()
train_id
```

```python
train_creative_id_embedding = pd.merge(train_features, creative_id_embedding, on='creative_id', how='left')
train_creative_id_embedding
```

```python
def extract_feat(df_, key, name, L, train_id):
    for i in range(L):
        t = df_.groupby(key).aggregate({'{0}_embe_{1}'.format(name, i):['mean','min','max']}).reset_index()
        t.columns = ['_'.join(j).strip('_') for j in t.columns.values]
        train_id = pd.merge(train_id, t, on=key, how='left')
        del t 
        gc.collect()
    return train_id
```

```python
train_creative_id_embedding_df= extract_feat(train_creative_id_embedding, 'user_id', 'creative_id', 32, train_id)
```

```python
train_creative_id_embedding_df
```

```python
train_creative_id_embedding_df.to_csv('output/features/train/creative_id_embedding_df.csv', index=False)
```

```python
del train_features, train_id, train_creative_id_embedding, train_creative_id_embedding_df
gc.collect()
```
